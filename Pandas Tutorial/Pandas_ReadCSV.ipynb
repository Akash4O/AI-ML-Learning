# Read Operation

```python
import pandas as pd

csv_1 = pd.read_csv("diabetes.csv")

csv_1.head()
```

#### Perticular Number of Rows Data

```python
csv_2 = pd.read_csv("diabetes.csv",nrows=3)

print(csv_2)
```

#### Perticular Column Data

```python
csv_3 = pd.read_csv("diabetes.csv",usecols=["Insulin","Age"])
# You can also give index number to retirve the column in usecols=[0] for Pregnancies column

csv_3
```

#### Skip A Row

```python
csv_4 = pd.read_csv("diabetes.csv",skiprows=[0])
#Just give the row index nummber in the skiprows=[]
csv_4
```

#### Index col Function

```python
csv_5 = pd.read_csv("diabetes.csv",index_col="BloodPressure")
#making BloodPressure column as the Index for the rows!
csv_5
```

#### To change the Header parameter

```python
csv_6 = pd.read_csv("diabetes.csv",header=4)
#updated Header parameter as the 4th row!
csv_6
```

#### changing the heading names parameter!

```python
csv_7 = pd.read_csv("diabetes.csv",names=["col1","col2","col3","col4","col5","col6","col7","col8","col9"])
#naming the columns with the new names!
csv_7
```

#### The Prefix Function

```python
csv_8 = pd.read_csv("diabetes.csv", )

csv_8
```

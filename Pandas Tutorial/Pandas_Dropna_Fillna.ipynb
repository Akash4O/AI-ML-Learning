## dropna and fillna to handle missing Data

```python
import pandas as pd

var = pd.read_csv('diabetes.csv')

var
```

### Drop missing Data

```python
#drops row of missing cells
var.dropna() 
```

```python
#drops column of missing cells
var.dropna(axis=1)
```

```python
#dropping the cell row where any value is null in a row

var.dropna(how = "any")
```

```python
#dropping the row where all values are null and ignores any other

var.dropna(how = "all")
```

```python
#subset parameter in dropna removing certain columns null values row!

var.dropna(subset = ["Age"])
```

```python
#inplace parameter : removes null data rows and creates a new dataset

#var.dropna(inplace=True)

#var
```

```python
#thresh paramete used to remove the specific null values row !
#var.dropna(thresh=3)
```

## Fillna

```python
#fillna can be used to fill the null vales with the specific data

var1 = pd.read_csv('diabetes.csv')

#var1.fillna(0) #to all data's

var1.fillna({"Pregnancies":"Nigga","BloodPressure":"Kaalu"}) #To fill values according to the columns
```

```python
#We can fill null value with either farwardfill{means filling with next data} obj.ffill()
#Backwardfill{means filling with previous data} obj.bfill()
var1.bfill()
```

```python
var.fillna(12,inplace=True)

var
```

```python
var.fillna(12,limit=1)
```

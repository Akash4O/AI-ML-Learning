# Merge

```python
import pandas as pd

#to merge 2 dataSets there must be a key component which is present in both the datasets
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})

var2 = pd.DataFrame({"A":[1,2,3,5],"C":[13,14,15,16]})

pd.merge(var1,var2,on="A")#here A is the key component

#only the common data of keycomponent will be added to the Merged dataset!
```

```python
pd.merge(var1,var2,on="A",how = "left")
```

```python
pd.merge(var1,var2,on="A",how = "right")
```

```python
pd.merge(var1,var2,on="A",how = "outer",indicator = True)
```

```python
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","id"))
```

# Concat

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([5,6,7,8])

pd.concat([sr1,sr2])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4]})

d2 = pd.DataFrame({"A":[1,2,3,4],"C":[13,14,15,16]})

pd.concat([d1,d2],axis=1,join="inner")
```

```python
pd.concat([d1,d2],axis=0,keys=["d1","d2"])
```

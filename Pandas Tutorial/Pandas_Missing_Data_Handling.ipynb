# Replace

```python
import pandas as pd

var = pd.read_csv('diabetes.csv')

var
```

```python
#you can replace any value By anything!
var.replace(to_replace=11,value=1)
```

```python
#regex mainly used to replace the a single alphabet with someword!
var.replace(["A-Za-z",0],"python",regex=True)
```

```python
var
```

# Interpolate

```python
#this only works with numeric data it fills data according to previous and further data
var.interpolate()
```

##### methods of interpolate :{'linear','time','values','nearest','zero','slinear','quadratic','cubic','barycentric','krogh','polynomial','spline','piecewise_polynomial','from_derivative','pchip','akima'}

```python
#methods can be used to tell the way to fill the data using what parameter method of maths calculations to be used!

var.interpolate(method="linear")
```

```python
#inplace used to set the ouput data as the original data
var.interpolate(method="quadratic",inplace=True)

var
```

## Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np

'''
Syntax : 
x=[]
y=[]
plt.bar(x,y)
plt.show()
'''
```

```python
x = ["Python","C","C++","Java"]

y = [85,60,70,82]

plt.xlabel("Languages",fontsize=10)
plt.ylabel("No.",fontsize=10)
plt.title("Bar Chart",fontsize=15)

plt.bar(x,y)

plt.show()
```

```python
#Can modify width of bars
plt.bar(x,y,width=0.4)
```

```python
#Changing the colors of bars
plt.bar(x,y,color="r",width=0.5)
```

```python
#giving multiple colors for bars

plt.bar(x,y,color=["r","black","g"])
```

```python
#aligning the x parameters 
plt.bar(x,y,align="edge")
#align has two arguments edge and center, By default it is center
```

```python
#highlighting the edges of bars obj.bar(edgecolor="color")
plt.bar(x,y,color = ["y","b","m","g"],edgecolor="r",linewidth=5)
#the width of highlighted line can be controlled by "linewidth" paramete
```

```python
#linestyle can be used to change the line appearance 
plt.bar(x,y,color = ["y","b","m","g"],edgecolor="r",linewidth=5,linestyle=":",alpha=0.4)
#alpha parameter is used to lighten the appearance of color in the bar!
```

```python
#adding legend in the graph
plt.bar(x,y,color = ["y","b","m","g"],edgecolor="r",label="Popularity")
plt.legend()
plt.show()
```

```python
#Overlapped Multi-BarGraph
z = [20,30,40,50]
plt.bar(x,y,width=0.4,label="Popularity")
plt.bar(x,z,width=0.4,label="Popularity1")
plt.legend()
plt.show()
```

```python
#Multi-BarGraph separated graphs
p = np.arange(len(x))
width = 0.2
p1 = [ j+width for j in p]

plt.bar(p,y,width,label="Popularity")
plt.bar(p1,z,width,label="Popularity1")

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
#creating a horizantol bar graph
plt.barh(p,y,width,label="Popularity")
plt.barh(p1,z,width,label="Popularity1")

plt.yticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

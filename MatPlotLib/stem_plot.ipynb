# Stem Plot

```python
import matplotlib.pyplot as plt
```

```python
'''
Basic Structure:

x=[]
y=[]
plt.stem(x,y)
plt.show()
'''
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]
z=[3,4,1,2,6,5]
```

```python
plt.stem(x,y)
plt.show()
```

```python
#Style of Lines can be changed using "Linefmt" parameter

'''
Character   line style
" '-' "      Solid line
" '--' "     Dashes line
" '-,' "     Dash-dot line
" ':' "      Dotted line
'''
```

```python
plt.stem(z,y,linefmt=":")
plt.show()
```

```python
#to cahnge the color and symbol of the graph we use "markerfmt" parameter

plt.stem(z,y,markerfmt="r+",linefmt=":") # r+ means red color and + symbol
plt.show()
```

```python
#We can use bottom parameter to specify where the bottom line starts from
plt.stem(z,y,markerfmt="r+",linefmt=":",bottom=4) 
plt.show()
```

```python
#We can change the base line color using the "basefmt" parameter and showning the legend
plt.stem(z,y,markerfmt="r+",linefmt=":",bottom=4,basefmt="g",label="Stem Chart",orientation="horizontal")
plt.legend()
plt.show()
```

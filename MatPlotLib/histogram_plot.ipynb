# Histogram Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

'''
x=[]
y=[]
plt.hist(x,y)
plt.show()
'''

x = list(np.random.randint(10,60,(50)))
```

```python
np = x

plt.hist(np,color="r",edgecolor="black")

plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No")
plt.show()
```

```python
#the data has been split between each the period so let's combine it by bins parameter

l=[10,20,30,40,50,60]
plt.hist(np,color="r",edgecolor="black",bins=l)

#the data of between 10 and 20 is being counted as one
# we can also use bins as "Auto"
```

```python
#spceifing the range parameter in histogram ,(0,100) is range in below code:

plt.hist(np,"auto",(0,100),edgecolor="black")
plt.xlabel("Range from 0 to 100")
plt.show()
```

```python
#cumulative parameter if set to -1 it plots diagram in reverse order
plt.hist(np,color="r",edgecolor="black",bins=l,cumulative=-1)
plt.show()
```

```python
#the bottom is starting from 0 in above graph we can start it from specific value, for ex: 10 in below graph
plt.hist(np,edgecolor="black",bottom=10)
```

```python
#we can align xvalues in 3 ways : right , mid , left
plt.hist(np,bins=l,edgecolor="black",align="mid")
plt.show()
```

```python
#changing the type of the histogram by histtype parameter , by default it is barchart 
plt.hist(np,bins=l,edgecolor="black",histtype="step")#setting histogram type as step instead of barchart
#the types can be bar,step,stepfilled,barstacked
```

```python
#to change it to Horizontal we use Orientation parameter , by default it is verticle
plt.hist(np,bins=l,edgecolor="black",orientation="horizontal")
```

```python
#we can adjust the width of the bars in the histogram by rwidth parameter
plt.hist(np,edgecolor="black",rwidth=0.8) # 0.2 out of 1 width has been removed
```

```python
#we can convert the histogram into log values or log table for example : log table
plt.hist(np,bins=l,edgecolor="black",log=True)
plt.show()
```

```python
#adding a legend in graph
plt.hist(np,bins=l,edgecolor="black",align="mid",label="python")
plt.legend()
plt.show()
```

```python
#splitting the graph by a line , for this we use axvline parameter : plt.axvline(point where line to be added,color of line)
plt.hist(np,bins=l,edgecolor="black",align="mid",label="python")
plt.axvline(35,color="r",label="Devider")
plt.legend()
plt.show()
```

```python
#to show the grid of the map we use grid function : plt.grid()
plt.hist(np,bins=l,edgecolor="black",align="mid",orientation="horizontal")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40] #values for graph
y=["C","C++","Java","Python"] # label for the values
```

```python
#showing piechart with labels
plt.pie(x,labels=y)
plt.show()
```

```python
#To highlight the specific part of data of make a part of data look like detached, for this we use explode parameter
ex=[0.2,0.0,0.0,0.0] #the difference from origin for detachment
plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
#changing the colors of the piechart
c = ["r","orange","m","b"]
plt.pie(x,labels=y,explode=ex,colors=c)
plt.show()
```

```python
#To show the percentage of each parameter in the graph we use parameter autopct="%0.1f%%"  in this  0.1 indicate 0's after the point
#and f indicates to the actual float value or percentage value
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%")
plt.show()
```

```python
#adding a shadow to the graph to make it look like a 3D graph and visually appealing 
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True)
plt.show()
```

```python
#To increase the radius of pie chart we can use radius parameter to increase or decrease the size of the piechart
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",radius=1.5)
plt.show()
```

```python
#We can also specify the distance between  labels and the graph by labeldistance parameter if <1 value it goes inside the graph
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",labeldistance=0.8)
plt.show()
```

```python
#we can change the degree of begining for example by default is starts from 0 degree we can change it any specific value
#we cna achieve this by startangle parameter

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",startangle=90)
plt.show()
```

```python
#We can change any text specific attributes or styles by textprops parameter it is given as dictionary input
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",textprops={"fontsize":15,"fontfamily":"Times new roman"})
plt.show()
```

```python
#The graph is showing elements in the counter clock-wise we can show it in clock-wise by setting Counterclock=False
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",counterclock=False)
plt.show()
```

```python
#To manipulate edge related propertie we use wedgeprops
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,wedgeprops={"linewidth":4,"edgecolor":"gray"})
plt.show()

#linewidth for changing the width of the shadow
#edgecolor for changing the color of the shadow
```

```python
#To change the placing of Orgiin or center of the graph we use center parameter and give the points(x,y) as the input
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",center=(3,3))
plt.show()
#the origin or center moved little to Java section and Java lable came inside the color or its section
```

```python
#To rotate the label we set the rotatelabel value as true
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",rotatelabels=True)
plt.show()
```

```python
#showing title and labels

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%")

plt.title("The Prog. Language Knowledge graph :")
plt.legend(loc=2)#we can change its position by loc paramete in legend
plt.show()

```

```python
#Creating the Dot Pie Chart means complete circle
plt.pie([1])
plt.show()
```

## Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

'''
basic structur:
x=[]
y=[]
plt.scatter(x,y)
plt.show()
'''
```

```python
c = pd.read_csv("diabetes.csv",usecols=["Pregnancies"],index_col=None)

temp = c.to_numpy().reshape(-1)

x= list(temp)

x
```

```python
c1 = pd.read_csv("diabetes.csv",usecols=["BloodPressure"],index_col=None)

y = list(c1.to_numpy().reshape(-1))

y
```

```python
c2 = pd.read_csv("diabetes.csv",usecols=["Insulin"],index_col=None)

y1 = list(c2.to_numpy().reshape(-1))

y1
```

```python
colors=y
sizes = [220,150,500,320,210,240,420,620,750]
plt.scatter(x,y,c=colors,edgecolor="black",s=sizes,cmap="BrBG",alpha=0.4)
plt.colorbar()
plt.xlabel("No. of Pregnancies")
plt.ylabel("Blood Preessure")
plt.title("BP and Pregnancies")
plt.show()
```

```python
#changing the shape of the scatter graph points by "marker"

plt.scatter(x,y,c=colors,edgecolor="black",s=sizes,cmap="viridis",alpha=0.4)
plt.scatter(x,y1,c=colors,edgecolor="black",s=sizes,cmap="viridis",marker="*",alpha=0.4)
plt.colorbar().set_label("Color_Bar")
plt.xlabel("No. of Pregnancies")
plt.ylabel("Blood Preessure")
plt.title("BP and Pregnancies")
plt.show()
```
